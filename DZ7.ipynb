```python
import threading
import time

def formula1(x):
    result = x ** 2 - x ** 3 + x ** 4 - x ** 5 + x + x
    return result

def formula2(x):
    result = x + x
    return result

def formula3(result1, result2):
    result = result1 + result2
    return result

def calculations_10000():
    start_time = time.time()

    result1 = []
    result2 = []

    def thread_formula1():
        for x in range(10000):
            res = formula1(x)
            result1.append(res)

    def thread_formula2():
        for x in range(10000):
            res = formula2(x)
            result2.append(res)

    thread1 = threading.Thread(target=thread_formula1)
    thread2 = threading.Thread(target=thread_formula2)
    
    thread1.start()
    thread2.start()

    thread1.join()
    thread2.join()

    result3 = formula3(sum(result1), sum(result2))

    end_time = time.time()
    duration = end_time - start_time

    print("10000 итераций:")
    print("Вычисления по формуле 1 заняли: %.3f сек" % duration)
    print("Вычисления по формуле 2 заняли: %.3f сек" % duration)
    print("Вычисления по формуле 3 заняли: %.3f сек" % (end_time - start_time))
    print("Общая длительность вычислений: %.3f сек" % duration)

def calculations_100000():
    start_time = time.time()

    result1 = []
    result2 = []

    def thread_formula1():
        for x in range(100000):
            res = formula1(x)
            result1.append(res)

    def thread_formula2():
        for x in range(100000):
            res = formula2(x)
            result2.append(res)

    thread1 = threading.Thread(target=thread_formula1)
    thread2 = threading.Thread(target=thread_formula2)

    thread1.start()
    thread2.start()

    thread1.join()
    thread2.join()

    result3 = formula3(sum(result1), sum(result2))

    end_time = time.time()
    duration = end_time - start_time

    print("\n100000 итераций:")
    print("Вычисления по формуле 1 заняли: %.2f сек" % duration)
    print("Вычисления по формуле 2 заняли: %.2f сек" % duration)
    print("Вычисления по формуле 3 заняли: %.2f сек" % (end_time - start_time))
    print("Общая длительность вычислений: %.2f сек" % duration)

calculations_10000()
calculations_100000()
```
